```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1440])

for i in range(0, 2100):
    img = mpimg.imread("C:\\Users\\Parth\\Downloads\\613\\UCMerced_LandUse\\Images\\dataset\\u"+str(i)+".jpg")
    gray = rgb2gray(img)
    plt.savefig("C:\\Users\\Parth\\Downloads\\613\\UCMerced_LandUse\\Images\\grayscale_dataset\\u"+str(i)+".jpg")
```

```python
# Importing Image and ImageOps module from PIL package
from PIL import Image, ImageOps
count = 0
for i in range(0, 2100):
    # creating an og_image object
    og_image = Image.open("C:\\Users\\Parth\\Downloads\\613\\UCMerced_LandUse\\Images\\dataset\\u"+str(i)+".jpg")

    # applying grayscale method
    gray_image = ImageOps.grayscale(og_image)
    
    if gray_image.size == (256, 256):
        count += 1
        gray_image.save("C:\\Users\\Parth\\Downloads\\613\\UCMerced_LandUse\\Images\\gray_scale_dataset\\u"+str(i)+".jpg")
print (count)
```

```python
from PIL import Image  
import PIL
from os import walk
import sys
import os

import numpy as np
import cv2
import random

def sp_noise(image,prob):
    row,col = image.shape
    gauss = np.random.randn(row,col)
    gauss = gauss.reshape(row,col)        
    noisy = image + image * gauss
    return noisy

init_path = "C:\\Users\\Parth\\Downloads\\613\\UCMerced_LandUse\\Images\\grayscale_dataset\\" + "\\"
dest_path = "C:\\Users\\Parth\\Downloads\\613\\UCMerced_LandUse\\Images\\speckle_dataset\\"  + "\\"
print(os.getcwd())
for (dirpath, dirnames, filenames) in walk(init_path):
    print(os.getcwd())
    for data_file in filenames:
        # print(os.getcwd())
        # print(data_file)
        image = cv2.imread(init_path+data_file,cv2.IMREAD_GRAYSCALE)
        # print(image) # Only for grayscale image
        noise_img = sp_noise(image,0.05)
        if noise_img.shape == (256,256):
            cv2.imwrite(dest_path+data_file, noise_img)
            print (data_file)
```
